```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import numpy as np
import statistics as st
import math

from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import stats
from scipy.stats import t
from scipy.stats import f
```

```python
sample_data=[1208,812,1217,1323,1308,1515,831,1021,1387,951,930,730]
s_mean=st.mean(sample_data)
s_sd=st.stdev(sample_data)
n=len(sample_data)
print('Sample Mean '+str(s_mean))
print('Sample Standard Deviation '+str(s_sd))
print('Sample Size = '+str(n))
```

```python
p_mean = 1272 #Population Mean
s_mean=1102.75 # Sample Mean
s_sd = 256.27 # Sample Standard Deviation
n = 12 # Sample Size
dof = n-1 # Degrees of Freedom
alpha = 0.05 #significance level
```

```python
t_statistics = (s_mean - p_mean) / (s_sd/math.sqrt(n)) # n = sample size 
print("The t-statistics is ", t_statistics)

neg_t_critical = t.ppf(alpha/2, dof) # dof to calculate t-critical values
print("The neg t-critical value is "+str(neg_t_critical))

pos_t_critical = t.ppf(1-alpha/2, dof) # dof to calculate t-critical values
print("The pos t-critical value is "+str(pos_t_critical))
```

```python
p_value = t.sf(abs (t_statistics), dof)*2  # sf- survival function
print(p_value)
```

```python
if neg_t_critical<t_statistics<pos_t_critical and p_value>alpha: 
    print("Accept Null Hypothesis i.e. no significant difference")
else:
    print("Reject Null Hypothesis i.e. significant difference")
```

```python
# For two Sample Tests

sample_IT=[4,5,6,4,4,3,2,5,3,4,7] 
sample_Finance=[2,3,4,5,4,7,6,4,3]

x1=st.mean (sample_IT)
x2=st.mean (sample_Finance)
v1=(st.stdev (sample_IT)) **2 # st.variance (sample_IT)
v2=(st.stdev (sample_Finance))**2 #st. variance (sample_IT)
n1=len (sample_IT)
n2=len (sample_Finance)
df= n1+n2-2
print("Sample mean of IT ratings =",x1)
print("Sample mean of Finance ratings = ",x2)
print("Sample variance of IT ratings =",v1)
print("Sample variance of Finance ratings =",v2)
print("Sample size of IT ratings = ", n1)
print("Sample size of Finance ratings = ",n2)
print('Degrees of freedom = ', df)
```

```python
n = abs (x1-x2) # numerator
d = math.sqrt(v1*(n1-1) + v2*(n2-1))*math.sqrt(1/n1 + 1/n2) #denominator
t_statistic=n/d
print('t-statistic =',t_statistic)

t_critical = t.ppf(1-0.025, df)
print('t-critical =',t_critical)

neg_t_critical = t.ppf(alpha/2, df)
print("\nThe neg t-critical value is "+str(neg_t_critical))

pos_t_critical = t.ppf(1-alpha/2, df)
print("The pos t-critical value is  "+str(pos_t_critical))
```

```python
p_value = t.sf(abs(t_statistic),df)*2
p_value
```

```python
if neg_t_critical<t_statistics<pos_t_critical and p_value>alpha: 
    print("Accept Null Hypothesis i.e. no significant difference")
else:
    print("Reject Null Hypothesis i.e. significant difference")
```
