```python
def feature_scale(scale):
  numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
  new_in_df = in_df_c.select_dtypes(include=numerics)
  if scale == 'minmax':
    scaler = MinMaxScaler()
  else:
    scaler = StandardScaler()
  df_scaled = pd.DataFrame(scaler.fit_transform(new_in_df.to_numpy()),columns=new_in_df.columns)
  return df_scaled
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

new_in_df = in_df_c.select_dtypes(include=numerics)
numeric_cols=new_in_df.columns
num_cols=4
n_bins = 40

# compute number of rows for plot 
num_rows= int(len(numeric_cols)/num_cols)+1

# setting canvas for plotting
fig, axs = plt.subplots(num_rows, num_cols, tight_layout=True,figsize=(20,10))

# plotting the numerical columns
for col, ax in zip(numeric_cols,axs.flatten()[:len(numeric_cols)]):
    ax.hist(new_in_df[col],bins=n_bins,density=True)
    ax.set_title(col)

plt.show()
```
